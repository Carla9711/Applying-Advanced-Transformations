```python
# import libraries
import pandas as pd
import json
```

```python
# load data
df_info = pd.read_csv('Data/superhero_info - superhero_info.csv')
df_powers = pd.read_csv('Data/superhero_powers - superhero_powers.csv')
```

```python
df_info.head()
```

```python
df_powers.head()
```

## Cleaning df_info

```python
# split Hero and Publisher
df_info['Hero|Publisher'].str.split('|', expand=True)
```

```python
df_info[['Hero','Publisher']] = df_info['Hero|Publisher'].str.split('|', expand=True)
df_info.head()
```

```python
df_info = df_info.drop(columns=['Hero|Publisher'])
```

```python
# convert measurements to dictionary and extract measures height and weight in separate columns
measure = df_info.loc[0,'Measurements']
print(type(measure))
measure
```

```python
# replace ' woith '
df_info['Measurements'] = df_info['Measurements'].str.replace("'",'"')
df_info['Measurements'].head()
```

```python
# convert to dictionary
df_info['Measurements'] = df_info['Measurements'].apply(json.loads)
```

```python
df_info['Measurements'].head()
```

```python
measure = df_info.loc[0,'Measurements']
print(type(measure))
measure
```

```python
# convert dictionary to dataframe with keys as column names
height_weight = df_info['Measurements'].apply(pd.Series)
height_weight
```

```python
# concatenate dataframes
df_info = pd.concat((df_info,height_weight), axis=1)
df_info.head()
```

```python
# remove units from Height and Weight columns
df_info['Height'] = df_info['Height'].str.replace(' cm','')
df_info['Weight'] = df_info['Weight'].str.replace(' kg','')
df_info.head()
```

```python
df_info['Height'].dtype
```

```python
df_info['Height'] = df_info['Height'].astype(float)
df_info['Height'].dtype
```

```python
df_info['Weight'].dtype
```

```python
df_info['Weight'] = df_info['Weight'].astype(float)
df_info['Weight'].dtype
```

```python
df_info.head()
```

```python
df_info = df_info.drop(columns='Measurements')
```

```python
df_info['Hero'].nunique()
```

## Clean df_powers

```python
df_powers.head()
```

```python
power = df_powers.loc[0,'Powers']
print(type(power))
power
```

```python
# convert string to list format
df_powers['Powers_list'] = '["'+df_powers['Powers'].str.replace(',','","')+'"]'
df_powers.head()
```

```python
# convert to list
df_powers['Powers_list'] = df_powers['Powers_list'].apply(json.loads)

df_powers['Powers_list'].head()
```

```python
power = df_powers.loc[0,'Powers_list']
print(type(power))
power
```

```python
# explode list
exploded = df_powers.explode('Powers_list')
exploded.head()
```

```python
cols_to_make = exploded['Powers_list'].unique()
cols_to_make
```

```python
# create new column for each power
for col in cols_to_make:
    df_col = df_powers.copy()
    df_col[col] = df_col['Powers'].str.contains(col)
    df_powers = pd.concat((df_powers,df_col[col]), axis=1)
df_powers.head()
```

```python
df_powers['hero_names'].nunique()
```

```python
df_powers = df_powers.drop(columns=['Powers','Powers_list'])
df_powers.head()
```

## Merge df_info and df_powers

```python
df_final = pd.merge(df_info, df_powers, left_on='Hero', right_on='hero_names')
```

```python
df_final.head()
```

```python
df_final.columns
```

```python
df_final = df_final.drop(columns='hero_names')
```

```python
df_final['Hero'].nunique()
```

Compare the average weight of super powers who have Super Speed to those who do not.

```python
df_final.groupby('Super Speed')['Weight'].mean()
```

The average weight of Heros who have Super Speed is 129.4 kg's while those that do not have Super Speed is 101.8 kg's
